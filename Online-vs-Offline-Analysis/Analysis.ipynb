```python
import pandas as pd
import matplotlib.pyplot as plt

df= pd.read_csv("online_vs_offline_learning_dataset.csv")
print(df.head())
print(df.info())

print(df.isnull(.sum()))
df.drop_duplicates(inplace=True)

analysis= df.grpupby("Learning_Mode")["Exam_scores"].mean()
print(analysis)


plt.bar(analysis.index, analysis.values, color=["skyblue","orange"])
plt.title("Average Exam score: Online vs Offline")
plt.ylabel("Average Score")
plt.show()

plt,scatter(df["study_hours"], df["Exam_score"],alpha=0.5,c="green")
plt.title("Study Hours vs Exam Scores")
plt.xlabel("Study Hours")
plt.ylabel("Exam Scores")
plt.grid(True)
plt.show()

plt.hist(df["Exam_score"],bins=10,edgecolor="black",color="red")
plt.title("Distribution of Exam Scores")
plt.show()
```
